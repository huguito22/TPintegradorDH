```python
import pandas as pd
#import geopandas as gpd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('properatti.csv')
```

```python
pf = pd.DataFrame(data)
```

```python
#Reducir el data set a cap fed
mask_capitalFederal = pf['state_name'] == 'Capital Federal'

dataset = pf[mask_capitalFederal]

dataset
```

```python
#Agrupamos por barrio
dataset_groupby_barrio = dataset.groupby('place_name')
```

```python
dataset_groupby_barrio
```

```python
#Completamos los valores nulos de price_usd_per_m2 con el promedio segun su barrio, usando transform y lambda.
price_m2_usd_filled = dataset_groupby_barrio['price_usd_per_m2'].transform(lambda grp: grp.fillna(grp.mean()))
```

```python
price_m2_usd_filled
```

```python
#Completamos los valores nulos de price_per_m2 con el promedio segun su barrio, usando transform y lambda.
price_m2_filled = dataset_groupby_barrio['price_per_m2'].transform(lambda grp: grp.fillna(grp.mean()))
price_m2_filled
```

```python
#Agregamos la columna de m2_usd_filled al dataset
dataset.loc[:, ['price_usd_m2_filled']] = price_m2_usd_filled
price_usd_fill = dataset['price_usd_m2_filled']
```

```python
#Agregamos la columna de m2_usd_filled al dataset
dataset.loc[:, ['price_m2_filled']] = price_m2_filled
price_m2_fill = dataset['price_m2_filled']
```
